# Visualizing World Cup Data With Seaborn

#### For this project you will be exploring data from the Fifa World Cup from 1930-2014 to analyze trends and discover insights about the world’s game, fútbol! This Fifa World Cup data is from Kaggle.

```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

matches = pd.read_csv('WorldCupMatches.csv')


#inspecting the datasets
print(matches.head(5))
# print(goals.head())
```

```python
#creating a new column 'Total Goals'
matches['Total Goals'] = matches['Home Team Goals'] + matches['Away Team Goals']
print(matches.head())
```

```python
#ploting a bar chart of total goals from world cup btw 1930-2014

fig, ax = plt.subplots(figsize=(12, 7))
sns.barplot(data=matches, x='Year', y='Total Goals')
sns.set_style("whitegrid")
sns.set_palette("pastel")
sns.set_context("poster", font_scale = .5, rc={"grid.linewidth": 0.6})

plt.xticks(rotation=45)
plt.title('Total Goals per World Cup',fontsize=16)
plt.legend()

plt.show()
plt.close()
```

```python
#ploting a box plot to visualize the distribution of the goals data

fig, ax2 = plt.subplots(figsize=(12, 7))
sns.boxplot(data=matches, x='Year', y='Total Goals')
sns.set_palette("Spectral")
sns.set_style("darkgrid")
sns.set_context("notebook", font_scale = 1.25, rc={"grid.linewidth": 0.6})

plt.xticks(rotation=30)
plt.title('Boxplot of Goals per Year of World Cups', fontsize=16)

plt.show()
```

```python
#ploting a bar chart of total goals per country
fig, ax = plt.subplots(figsize=(18, 12))
sns.barplot(data=matches, x='Away Team Initials', y='Total Goals')
sns.set_style("whitegrid")
sns.set_palette("pastel")
sns.set_context("notebook", font_scale = 1.0, rc={"grid.linewidth": 0.6})

plt.xticks(rotation=60)
plt.title('Total Goals of Teams at World Cup',fontsize=16)
plt.legend()
plt.savefig('Total Goals of teams at world cup')


plt.show()
plt.close()
```
